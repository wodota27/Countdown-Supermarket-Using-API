```python
StoreName = []
Street = []
Town = []
Postal_Code = []


for result in result_items:
    try:
        StoreName.append(result['site']['name'])
    except:
        StoreName.append('')
    
    try:
        Street.append(result['site']['addressLine1'])
    except:
            Street.append('')
    
    try:
        Town.append(result['site']['suburb'])
    except:
            Town.append('')
    
    try:
        Postal_Code.append(result['site']['postcode'])
    except:
        Postal_Code.append('')
    

    
```

```python
Countdown_df = pd.DataFrame({'Store Name':StoreName, 'Street':Street, 'Town and City':Town, 'Postal_Code': Postal_Code})
```
